2. Pass / Fail Analyzer

```python
# list of student marks
marks = [45, 78, 90, 33, 60]

pass_count = 0
fail_count = 0

# Checking each student's marks
for m in marks:
    if m >= 50:
        pass_count += 1
    else: 
        fail_count += 1
# Printing final result clearly

print("Final Result Summary")
print("-------------------")
print("Total Students:",len(marks))
print("Total Passed Students:", pass_count)
print("Toatl Failed Students:", fail_count)
```

3. Simple Data Cleaner

```python
# Raw names list
names = ["Alice", "bob", "CHARLIE"]

cleaned_names = []

# Cleaning each name:
for name in names:
    name_without_spaces = name.strip()   # removes extra spaces
    final_name = name_without_spaces.lower()
    cleaned_names.append(final_name)

print("Cleaned Names List:", cleaned_names)
```

4. Message Length Analyzer

```python
# Messages List
messages = ["Hi", "Welcome to the platform", "OK"]

for msg in messages:
    length = len(msg)
    print(f"Message: '{msg}' | Length:{length}")

    if length > 10:
        print("This message is longer than 10 characters")
```

5. Error Message Detector

```python
# System logs
logs = ["INFO", "ERROR", "WARNING", "ERROR"]

error_count = 0

# Counting how many times "ERROR" appears
for log in logs:
    if log == "ERROR":
        error_count += 1
        
# Printing total error count
print("Total number of ERROR entries:",error_count)
```
